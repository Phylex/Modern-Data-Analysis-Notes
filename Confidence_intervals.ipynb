## Confidence Intervals

To determin the confidence interval $\mathrm{CI}$ we need a bunch of things all working together and playing nice with each other to get to an understandable result.

* A **test statistic** $t(\vec{x})$. The test statistic provides a measure for a property that we base our hypothesis on. This means that if we see the property that the test statistic is trying to express this is a (hopefully strong) indicator, that our hypothesis is right (well techically the it is a strong indicator that our hypothesis is not wrong (yet)). If we however don't see the property that should be expressed by the test statistic in the data, we can assume that our hypothesis is wrong. The test statistic is usually a function $t\colon \mathbb{R}^n \to \mathbb{R}$ to a codomain with a partial order (Here and most commonly the codomain is $\mathbb{R}$).
 
* A **model** $M(\vec{\theta})$ that describes the process/experiment in question. The model can be an implicit part of our hypothesis test when for example we want to test a specific parameter set, or can be an explicit part of our hypothesis test if we are for example checking what order of polynomial describes our data best.

* A **hypothesis**. As we can never know the truth, we need an assumption about our approximate truth. This then becomes our working assumption that the subsequent confidence interval and tests will be based on. The hypothesis generally includes a model and a specific choice of parameters to characterize the data. Generally when we are trying to calculate confidence intervals we have allready done basic sanity checks for our model with our specific choice of parameters and have come to the conclusion that they do in fact describe our experiment/process/data to a reasonable degree (we can do that by formally doing hypothesis tests or (and this is done far more often) using gut feeling and a reasonable explanation other humans generally agree with (this is the art and a bit of craftsmanship that make up the less formal part of the sciences)). 

* A **confidence level**. This is fairly self explanatory but I'll go over it just in case. The higher our confidence level, the surer we are that the statement we make with our model and parameters is true. This is different from being precice. We can be very vague and as a result the statement will be true, so that the resulting confidence level is high (close to or even equal to 1). I can for example state that the maximum speed of my dream car is anywhere from $0$ to $c$. This statement is definitely true for all cars (so the confidence level is 1), and as a result tells us fairly little about my dream car. If on the other hand I state that this awesome car has a maximum speed in the range of 220 km/h to 350 km/h you could for example exclude my current car (a Reno Megane from 2003) because it has a top speed in the range of 195 km/h (depending on various environmental parameters like temperature), however now there is the possibility that I am wrong (I have not stated any external conditions like the amount of fuel in the tank or battery charge level and with an empty tank the max speed of the car is obviously 0 km/h, so in fact that statement is not entirely true (as long as I don't start requireing external factors to have a specific value, like the car being fuled). So we could set our confidence level $\mathrm{CL}$ to 1, but this has the downside, that then we have to essentially include all models of every possible parameter set just to be sure (we are absolutely certain remember (see the example above). We can however approach this absolute certainty arbitrarily close without having to extend the $\mathrm{CL}$ to every single possible model and parameter set by refining our model and parameters through an increasingly large set of measurements and constraint parameters. (This effort becomes infinite while we approach $\mathrm{CL} = 1$) 

## Upper limit on confidence intervals

The upper limit on confidence interval of a parameter $\theta_i$ is a value that can be given for any observation $\vec{x}_{obs}$ in combination with an arbitrary confidence level $\mathrm{CL} = \alpha$. The Idea is that the Test statistic, being a function with the codomain $\mathbb{R}$, maps an arbitrarily complex multidimensional probability density onto a single orderd codomain. If we now compute $t(\vec{x}_{obs})$ we now have a "location" in an ordered set. We can now determin the Probability of taking any possible observation $\vec{x}_i$, compute $t(\vec{x}_i)$ and end up on the left (so $t(\vec{x}_i) < t(\vec{x}_{obs})$) or on the right ($t(\vec{x}_{obs}) \leq t(\vec{x}_i)$) of $\vec{x}_{obs}$.

The Probability density used to compute this measure is described by our (arbitrary) Model $M(\vec{\theta})$ and the test statistic $t(x)$ so we can now observe how this Density and changes when the parameter $\theta_i$ is changed. $\theta_i$ is a single value from the parameter vector $\vec{\theta}$.
A changed Model $M(\vec{\theta_1})$ also means that the ratio of possible obersvations $r_p = \frac{|\{\vec{x}_l\}|}{|\{\vec{x}_l\}|+|\{\vec{x}_r\}|}$ where $\{\vec{x}_l\}$ is the set of all possible obsevations with $t(\vec{x}_l) \lt t(\vec{x}_{obs})$ and $\{\vec{x}_r\}$ the set of all possibilities where $t(\vec{x}_{obs})\leq t(\vec{x}_r)$.

It is now possible to construct the upper limit. The upper limit is defined by the requirement that $r_p = \frac{|\{\vec{x}_l\}|}{|\{\vec{x}_l\}|+|\{\vec{x}_r\}|} \leq 1-\alpha$ with the added constraint that $r_p$ gets as close to $1 - \alpha$ as is possible by varying $\theta_i$.
So as a consequence of this we vary $\theta_i$ until we find the value in satisfying the requirements stated above.

This has maybe sounded a bit abstract so I'll now show a few examples to make things more concrete. It is also worth pointing out that in most useful cases $t(\vec{x})$ is defined in such a way that maps a contiguous volume $V$ in the observations space to a contiguous set of values in the codomain of $t(\vec{x})$. This makes it possible to express $|\{\vec{x}_l\}|$ and $|\{\vec{x}_r\}|$ as Integrals over $V_l$ and $V_r$ (these are the spaces of all obersations that are mapped by $t$ to values left of $t(\vec{x}_{obs})$ and values right of $t(\vec{x}_{obs})$ respectively) greatly simplifying matters.

## Interpretation of the upper limit
The upper limit is given for a Model parameter $\theta_i$. As such it is a statement about this parameter.

The confidence interval for a specific confidence level $\alpha$ is defined as the Volume of the parameter space of our model (of course this implicitly assumes that our model is correct and we only need to fit the parameters of that model to come to a true model of our observed process/property) for which we have at least the confidence $\alpha$ that the true value for $\vec{\theta}$ lies within this interval.

As such the upper limit of $(\theta_{i})_{UL}$ is the value of $\theta_i$ for which we leave that Volume in the direction of $\theta_i$.
This is of course only easily determined if the parameter of the model $\theta_i$ does not interfere with the other parameters of the model. In other words if the parameters are not correlated through the model.

## The lower limit
For the lower limit we now only have to change the definition of $r_p$ to $r_p = \frac{|\{\vec{x}_l\}|}{|\{\vec{x}_l\}|+|\{\vec{x}_r\}|} \geq 1-\alpha$ or to $r_p = \frac{|\{\vec{x}_r\}|}{|\{\vec{x}_l\}|+|\{\vec{x}_r\}|} \leq \alpha$ producing the lowest possible value for $\theta_i$ for which the model is still within the Cofidence interval (given that the parameters are not correlated).

## Examples

#### Example 1: The temperature of an oven.
Let's say we have an oven and can set the temperature of the oven using a knob.
Let's also assume we set the oven to a temperature of 200 C. After the oven has heated to that temperature
we take a measurement of that temperature using a very precise thermometer (so precice that we can simply ignore the uncertainty introduced by the thermometer). After the measurement we switch off the oven and let it cool until we repeat the cycle over and over again producing the following histogram.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats as st
from scipy import optimize as op

# we repeat the cycle of heating-> measuring->cooling a thousand times
spread = 10
temperature_measurements = np.random.randn(1000) * spread + 200
hist, bins, patches = plt.hist(temperature_measurements, edgecolor="black")
plt.show()
```

We can see that we generally measure a temperature around 200 C, which is good news because that is after all the temperature that we had set on the knob. We also see that some of the time the temperaure can be as high as 230 C and as low as 170 C (roughly).

So what do we do when I ask the inverse question:
> What temperature is the knob set to if I measure a temperature $T_0$?

First off, given what we saw, we can observe that there is no 1 to 1 correlation (bijective mapping) between knob positions and measured temperatures. (Otherwise we would have had measured 200 degrees for every iteration of the cycle) So it is not possible to give an exact knob position. However we can give a Range of possible values and we can give a confidence estimate for these ranges (Confidence intervals and Confidence levels).

To be able to do that let's first decide on a confidence level (it is arbitrary after all but any value will suffice. You can adapt the code to calculate the different numbers for your value). I'll go for 95%, so $\alpha = 0.95$. This is a fairly common confidence level. Sometimes you'll find others like 75% or even 50%.

```python
alpha = 0.95
```

Now we also need a model. As our measurement is a single value (the temperature $T$ of the oven) and we only have one parameter (the angle of the knob $\theta$) we can use a plain old function for our model.

If we look at the histogram above you could guess that (and you'll find that the code in deed confirms this suspicion) that the measured
temperatures follow a gaussian distribution with the mean 200. Let's therefore assume that the measured temperature follows a gausian distribution with the mean of that distribution being the best estimate for the knob position.

Please take note, that even though the code confirms this, in real life (so a situation where you can't just look at the code that the universe runs on) you can only have hypotheses. These hypotheses can be (and often are) formulated based on hunches and prior experience, in the end the initial hypothesis does not matter anyway (this hypothetical end is after the heat death of the universe but it can be shown that it is in deed possible to arrive at the right conclusion no matter how wrong you started out the only difference is in how quickly you arrive at the right one).

So we will use our intuition and choose the gaussian as our model.

```python
model = st.norm #scipy.stats allready implemets the gaussian with some cool extra features, so ill use that one
```

Now we could notice, that we have two free parameters for the gaussian (the mean $\mu$ and the standard deviation $\sigma$), but only one free parameter for our model (the angle $\theta$ of our knob). This means that we have to make an assumption about how the parameters $\sigma$ and $\mu$ depend from our knob angle $\theta$.

I'd say a reasonable assumption is that the standard deviation is some property of the electronics that control the oven temperature and therefor that $\sigma$ is not dependant on the $\theta$ but has a constant value for this particular oven. The mean temperature $\mu$ is hopefully the temperature that the knob is set to (otherwise the knob would not perform it's function of controlling the temperature of the oven after all). This leads us to the assumption that $\mu$ and $\theta$ are proportional to each other with a proportionality constant. The next step is then to determin the proportionality constant.

So a full turn of the knob is a rotation by $2\pi$. Let's assume that the oven can be set anywherer from 0$^\circ$ C to 350$^\circ$ C. Now let's assume that one full rotation of the knob sets the oven to it's maximum temperature. Then we have our proportionality constant through the formula

$$\frac{350^\circ \text{C}}{2\pi} = 55.70423^\circ \text{C per radian} $$

To get a first guess for $\sigma$ we can use our model and try to fit the model to the data using a fitting algorithm.

```python
(mu, sigma) = model.fit(temperature_measurements)
print("mu = {}".format(mu))
print("sigma = {}".format(sigma))
```

Now we have a first estimate for the standard deviation with `sigma`. (If we peak into the code that generated the data we can find that we are in deed pretty close to it's original value).

With this we can make our model a bit more specific, as from now on, our specific model can't be applied to any oven but only this one. Implicit in the specification was, that the standard deviation was dependent on the electronics in each oven, and as the electronics (or control software) inside the oven could be different for different ovens, we can only apply the specific model to the data produced by the specific oven.

```python
specific_model_pdf = lambda t, mu: model.pdf(t, loc=mu, scale=sigma)
specific_model_cdf = lambda t, mu: model.cdf(t, loc=mu, scale=sigma)
specific_model_sf = lambda t, mu: model.sf(t, loc=mu, scale=sigma)
specific_model_isf = lambda t, mu: model.isf(t, loc=mu, scale=sigma)
```

Now just let us check that in deed the model describes the data we have allready measured

```python
x = np.linspace(min(bins), max(bins), 1000)
plt.hist(temperature_measurements, edgecolor='black', density=True)
plt.plot(x, specific_model_pdf(x, 200))
plt.show()
```

That looks good.

Now to that we have constructed a model and performed basic sanity checks we can start calculating our confidence intervals as well as upper and lower limits.

Note that I have not discussed the hypothesis yet, and still I am claiming that I can now go ahead and calculate the confidence level. This is in part due to the simple nature of the example and in part because I have not made the hypothesis explicit yet (we are working with hypotheses after all). Our Model is the hypothesis. I stated that we can't ever really know what the truth is and I said that I assume that the electronics are responisble for the spread of the temperatures of the oven (that's an implicit hypothesis right there) or that the knob angle and the temperature are proportional (another hypothesis). For this hypothesis we however did do a sanity check turning the knob to 3.59 radian counter clockwise (assuming a mathematical positive rotation increases temperature) and measured the temperature. The temperature then also checked out to be 200 C which means that we have a basic sanity check for this assumption and can verify the proportionality constant of roughly $55.70423^\circ \text{C per radian}$.

So really we are testing all these hypotheses simultaneously and it is the job of the experimentor to either use intuition or (in the more complicated cases more rigorous methods) to verify these hypotheses (and depending where you work a quantifiable measure of the validiy of the hypothesis is required)). With that out of the way we can turn our attention to the test statistic.

The test statistic $t(\vec{T})$ in this case is fairly trivial. Our Model allready has a [partially ordered](https://en.wikipedia.org/wiki/Partially_ordered_set) codomain (our model spits out probabilities to get a temperature given a knob angle input, and temperatures are real numbers and real numbers are [totally ordered](https://en.wikipedia.org/wiki/Total_order) which includes partial order). The test statistic simply incorporates the proportionality constant between knob angle and the expected temperature of the oven.

```python
def t(T):
    return (2*np.pi)/350 * T
```

So now to the actual point of this chapter, the determination of the confidence interval with the corresponding upper and lower limit.

The upper limit is calculated for a set of measurements $\vec{T}$ (I will start with a single measurement $T_0$ for now) and is the highest value of $\theta$ for which the measured value still lies inside the confidence interval specified by `alpha`.

```python
# plot the data and the model
def plot_complete(ax, measurements, temp, alpha, hist_color, model_color, upper_lim_color, lower_lim_color):
    hist, bins, patches = ax.hist(temperature_measurements, edgecolor='black', density=True, alpha=0.5)
    x = np.linspace(min(bins), max(bins), 1000)
    ax.plot(x, specific_model_pdf(x, temp), color='orange', label=r'model for {}$^\circ$C'.format(temp))
    
    lower_limit = specific_model_isf(alpha, temp)
    upper_limit = specific_model_isf(1-alpha, temp)
    if upper_lim_color == lower_lim_color:
        ax.vlines(x=lower_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=lower_lim_color, label='limits of {}% of measurements'.format(alpha*100))
        ax.vlines(x=upper_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=upper_lim_color)
    else :
        ax.vlines(x=lower_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=lower_lim_color, label='lower limit of {}% of measurements'.format(alpha*100))
        ax.vlines(x=upper_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=upper_lim_color, label='upper limit of {}% of measurements'.format(alpha*100))  
    fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=len(measurements), fancybox=True, shadow=True)
    plt.rc('axes', axisbelow=True)
    ax.grid()
    plt.show()

x = np.linspace(min(bins), max(bins), 1000)
plt.hist(temperature_measurements, edgecolor='black', density=True, alpha=0.5)
plt.plot(x, specific_model_pdf(x, 200), color='orange', label=r'model for {}$^\circ$C'.format(200))

### now use the model to determin the upper and lower limit
# this is the inverse survival function specified for our oven so as a result we get the temperature
# for which the probability of measuring a larger value is 95%, inverse because we give it a probability and get
# a temperature. The non inverted function expects a temperature and returns the corresponding probability of measuring a
# larger temperature
lower_limit = specific_model_isf(alpha, 200)
upper_limit = specific_model_isf(1-alpha, 200)

# draw vertical lines for the upper and lower limits
plt.vlines(x=lower_limit, ymin=0, ymax=max(specific_model_pdf(x, 200))+0.01, color="cyan", label='lower limit of {}% of measurements'.format(alpha*100))
plt.vlines(x=upper_limit, ymin=0, ymax=max(specific_model_pdf(x, 200))+0.01, color="navy", label='upper limit of {}% of measurements'.format(alpha*100))

# make the plot a little nicer
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=1, fancybox=True, shadow=True)
# this line puts the grid behind the plotted elements
plt.rc('axes', axisbelow=True)
plt.grid()
plt.show()
```

Now let's assume we measure a temperature $T_0$ of 192$^\circ$C and want to figure out the range of knob positions that could produce $T_0$ that lies within 95% of values around the corresponding mean. As we are looking for the limit of that range we want to know the temperatures set by the knob where $T_0$ lies exactly on the limit temperature of the intervals for the knob positions that constitute the upper and lower limit of the 95% Confidence Interval.

This sounds very confusing (it was for me too) but a 

```python
#### generate a set of temperature measurements for different knob positions.

# first we start with arbirary knob positions from 0 to 2*pi 
knob_positions = [1, 2, 3, 5, 2*np.pi] 

# translate the knob positions into corresponding temperatures
def knob_to_temp(k):
    return 350/(2*np.pi) * k
temps = [knob_to_temp(knob_pos) for knob_pos in knob_positions]

# generate the measurements
measurements = [np.random.randn(1000) * spread + mean_temp for mean_temp in temps]
```

```python
# define a function that takes care of the plotting. it is based on the previous code
def plot_complete(ax, measurements, temp, alpha, hist_color, model_color, upper_lim_color, lower_lim_color):
    hist, bins, _ = ax.hist(measurements, edgecolor='black',color=hist_color, density=True, alpha=0.7)
    x = np.linspace(min(bins), max(bins), 1000)
    ax.plot(x, specific_model_pdf(x, temp), color=model_color, label=r'model for knob position of {:.0f}$^\circ$C'.format(temp))
    
    lower_limit = specific_model_isf(alpha, temp)
    upper_limit = specific_model_isf(1-alpha, temp)
    if upper_lim_color == lower_lim_color:
        ax.vlines(x=temp, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color='gray', linestyle='--')
        ax.vlines(x=lower_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=lower_lim_color, label='{:.0f}% limit for {:.0f}$^\circ$C'.format(alpha*100, temp))
        ax.vlines(x=upper_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=upper_lim_color)
    else :
        ax.vlines(x=temp, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color='lightgray', linestyle='--')
        ax.vlines(x=lower_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=lower_lim_color, label='lower limit of {:.0f}% of measurements'.format(alpha*100))
        ax.vlines(x=upper_limit, ymin=0, ymax=max(specific_model_pdf(x, temp))+0.01,
                   color=upper_lim_color, label='upper limit of {:.0f}% of measurements'.format(alpha*100))
    return upper_limit, lower_limit
    
```

```python
#### plot the different measurements into the same plot
fig= plt.figure(figsize=(16,4))
ax= fig.add_axes([0.1,0.1,0.9,0.9])
hist_colors = ['darkblue', 'blue', 'royalblue', 'deepskyblue', 'darkturquoise']
limit_colors = ['darkgreen', 'green', 'lime', 'greenyellow', 'yellowgreen'][::-1]
model_colors = ['red', 'firebrick', 'coral', 'orange', 'gold'][::-1]
for hist_color, model_color, limit_color, m, temp in zip(hist_colors, model_colors, limit_colors, measurements, temps):
    plot_complete(ax, m, temp, alpha, hist_color, model_color, limit_color, limit_color)
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=len(measurements), fancybox=True, shadow=True)
# this line puts the grid behind the plotted elements

plt.rc('axes', axisbelow=True)
plt.grid()
plt.show()
```

If we perform the experiment with different knob positions we get different ranges of measurements where the histogram of these measurements will show the characteristic bell curve of a gaussian distribution centered around the corresponding temperature that we set the knob to.

Now we want to find the range of possible knob positions for a $T_0$ of 192$^\circ$C.

```python
t0 = 192
fig= plt.figure(figsize=(16,4))
ax= fig.add_axes([0.1,0.1,0.9,0.9])
ax.vlines(x=t0, ymin=0, ymax=0.05, label=r'$T_0$ = {}'.format(t0), color='black')
plot_complete(ax, measurements[2], temps[2], alpha, 'blue', 'coral', 'greenyellow', 'greenyellow')
plot_complete(ax, measurements[3], temps[3], alpha, 'deepskyblue', 'firebrick', 'green', 'green')
plot_complete(ax, temperature_measurements+5, 205, alpha, 'navy', 'gold', 'lime', 'lime')
plt.grid()
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=len(measurements), fancybox=True, shadow=True)
plt.show()
```

As I hope it is fairly obvious by now we can see that the black line at 192$^\circ$C is just inside the 95% interval for the measurements with a knob set to 3.6801 radians which corresponds to 205$^\circ$C but outside the intervals for the knob positions at 3 and 5 radians respectively. So we need to figure out how far we can shift the distribution up and down and still have $T_0$ end up inside the interval for the knob setting.

To figure this out in this special case we can use our knowledge of the Gaussian distribution to figure out the distance in $T$ of the limits to the knob position, as everything gets shifted by the same amount when we turn the knob (in other words the distribution is independant from the knob position). So let's figure out the distance.

```python
# we have a model for our specific oven we use and we can use the isf (see code section above)
knob_pos = 3
CL = 0.95
temp = knob_to_temp(knob_pos)
dist_to_upper_limit =  temp - specific_model_isf(CL, temp)
dist_to_lower_limit = temp - specific_model_isf(1-CL, temp)
print(dist_to_lower_limit)
print(dist_to_upper_limit)
```

If we actually have the `isf` the task is a fairly simple one as we can use it to get the position in $T$ of the upper and lower limit and simply subtract the set temperature from that value(assuming that the mean and the temperature set by the knob coincide which is the case here we can also use the mean). Try out different knob positions just to be sure that the difference stays the same for all positions, will however change for different confidence level.

```python
def calc_limits(CL, temp):
    dist_to_upper_limit = temp -specific_model_isf(CL, temp)
    dist_to_lower_limit = temp -specific_model_isf(1-CL, temp)
    return dist_to_upper_limit, dist_to_lower_limit
```

```python
# we need to generate data at the point t_0
t0dist = np.random.randn(1000) * spread + t0
fig= plt.figure(figsize=(16,4))
ax= fig.add_axes([0.1,0.1,0.9,0.9])
dTul, dTll = calc_limits(0.95, t0)
plot_complete(ax, t0dist+dTll, t0+dTll, alpha, 'blue', 'coral', 'greenyellow', 'greenyellow')
plot_complete(ax, t0dist+dTul, t0+dTul, alpha, 'deepskyblue', 'firebrick', 'green', 'green')
ax.vlines(x=t0, ymin=0, ymax=0.05, label=r'$T_0$ = 192', color='black')
plt.grid()
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=3, fancybox=True, shadow=True)
plt.show()
```

As it can be seen now, $T_0$ is exactly at the upper end of the interval of temperatures inside the 95% interval of the knob position that constitutes the lower limit of the Confidence Interval with a confidence level of 95%. Similarly $T_0$ is exactly at the low end of the temperatures insider the 95% range of the knob position that constitutes the upper limit of the Confidence interval with confidence level of 95%.

This is the confusion I mean. $T_0$ is at the upper end of the interval for the lower limit and vice versa.

As you might have noticed we have calculated temperatures and not knob positions. We are able to do this because the test statistic $t(x)$ is linear and bijective so we can iditify the upper limit temperature to a knob position.

```python
upper_most_knob_setting = t(t0+dTul)
lowest_most_knob_setting = t(t0+dTll)
print(upper_most_knob_setting)
print(lowest_most_knob_setting)
```
